```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
# total size of the target dataset
N = 20
#train size
Ntrain = 12
X_ = np.linspace(0,2*np.pi,N).reshape(-1,1)
Y_ = np.sin(X_)
```

```python
plt.plot(X_,Y_)
```

```python
#打乱内部顺序
X,Y = shuffle(X_,Y_)
```

```python
# 把数据集分成训练集和测试集
Xtrain = X[:Ntrain]
Ytrain = Y[:Ntrain]
Xtest = X[Ntrain:]
Ytest = Y[Ntrain:]
```

```python
#训练
model = KNeighborsRegressor(n_neighbors = 5, weights = 'distance')
model.fit(Xtrain, Ytrain)
Ypredict = model.predict(Xtest)
```

```python
plt.scatter(Xtrain, Ytrain, s=50, alpha = 0.7, c='blue')
plt.scatter(Xtest, Ytest, s=50, alpha = 0.7, c='red')
plt.scatter(Xtest, Ypredict, s=50, alpha = 0.7, c='orange')
```
