```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
# Generate data
N = 100
D = 2
X = np.random.randn(N,D)
X[:50] += np.array([1,1])
X[50:] += np.array([-1,-1])
Y = np.array([0]*50+[1]*50)
# 数据可视化
plt.scatter(X[:,0],X[:,1],c=Y,alpha =0.7) 
```

```python
# 在图上画出实际的分类边界，为此我编写函数，绘出边界
def plot_boundary(x,model):
    plot_step = 0.02
    x_min, x_max = min(x[:,0])-1, max(x[:,0])+1
    y_min, y_max = min(x[:,1])-1, max(x[:,1])+1
    xx,yy = np.meshgrid(np.arange(x_min,x_max, plot_step), np.arange(y_min, y_max, plot_step))
    z = model.predict(np.hstack((xx.ravel().reshape(-1,1), yy.ravel().reshape(-1,1))))
    z = z.reshape(xx.shape)
    plt.contour(xx, yy, z, cmap = plt.cm.Paired)
```

```python
#实际训练并可视化
# knn 1
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X, Y)

plt.scatter(X[:,0], X[:,1], s=50, c=Y, alpha=0.7)
plot_boundary(X, model)
plt.show()

# knn 2
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X, Y)

plt.scatter(X[:,0], X[:,1], s=50, c=Y, alpha=0.7)
plot_boundary(X, model)
plt.show()

# knn 3
model = KNeighborsClassifier(n_neighbors=5,weights='distance')
model.fit(X, Y)

plt.scatter(X[:,0], X[:,1], s=50, c=Y, alpha=0.7)
plot_boundary(X, model)
plt.show()
```
